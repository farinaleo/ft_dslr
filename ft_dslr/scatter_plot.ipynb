```python
import pandas
import json
import pandas as pd
import matplotlib.pyplot as plt

main_column = 'Hogwarts House'
with open('setup.json') as f:
    data = json.load(f)
path = data["scatter_path"]
print(f'path : {path}')
```

```python
def csv_to_dataframe(csv_file: str) -> pandas.DataFrame:
    """Converts a CSV file into a Pandas dataframe"""
    dtf = pandas.read_csv(csv_file)
    return dtf
csv_to_dataframe(path)
```

```python
def scatter_matrix_plot(dtf: pd.DataFrame, x_feature=None, figsize=(20,20)):
    """Plot a scatter matrix for each course."""
    dtf = dtf.drop(columns='Index', inplace=False)
    nb_features = dtf.select_dtypes(include=['number']).columns.tolist()
    
    # remove the 'Hogwarts House' if it is found in the course list
    if main_column in nb_features:
        nb_features.remove(main_column)

    nb_features.reverse()
    feature_len = len(nb_features)
    if x_feature is None:
        x_feature = nb_features
    x_feature_len = len(x_feature)
    
    font = {'weight': 'light', 'size': 7}
    plt.rc('font', **font)

    # Set up the matrix (subplot)
    figure, axes = plt.subplots(nrows=x_feature_len, ncols=feature_len, squeeze=False, figsize=figsize, sharex=False, sharey=False)

    # Loop on each course (on x and y) to create the matrix
    for x in range(0, x_feature_len):
        for y in range(0, feature_len):
            # Remove the scatter if a course is compared with itself
            if x_feature[x] != nb_features[y]:
                plot_scatter(dtf[[main_column, x_feature[x], nb_features[y]]], axes[x, y], x_feature[x], nb_features[y])
            # Add scales only for the first column and the last row
            if x != x_feature_len - 1:
                axes[x, y].tick_params(labelbottom=False)
                axes[x, y].xaxis.set_ticks_position('none')
            else:
                axes[x, y].set_xlabel(nb_features[y].replace(' ', '\n'))
            if y != 0:
                axes[x, y].tick_params(labelleft=False)
                axes[x, y].yaxis.set_ticks_position('none')
            else:
                axes[x, y].set_ylabel(x_feature[x].replace(' ', '\n'))
            axes[x, y].tick_params(labelrotation=45.0)
            axes[x, y].tick_params(labelright=False)
            axes[x, y].tick_params(labeltop=False)
            axes[x, y].spines['right'].set_visible(False)
            axes[x, y].spines['top'].set_visible(False)

    plt.legend(['G', 'H', 'S', 'R'], loc='center left', bbox_to_anchor=(1, 0.5))
    plt.show()
```

```python
def plot_scatter(dtf: pd.DataFrame, axe, x_name: str, y_name: str):
    """Create and plot a scatter to compare two courses (x_name and y_name)"""
    # Extract each house
    _gryffindor = dtf[dtf[main_column] == 'Gryffindor']
    _hufflepuff = dtf[dtf[main_column] == 'Hufflepuff']
    _slytherin = dtf[dtf[main_column] == 'Slytherin']
    _ravenclaw = dtf[dtf[main_column] == 'Ravenclaw']

    # Plot each house one by one to apply the associate color
    axe.scatter(_gryffindor[y_name], _gryffindor[x_name], s=2, color='red', alpha=0.5)
    axe.scatter(_hufflepuff[y_name], _hufflepuff[x_name], s=2, color='yellow', alpha=0.5)
    axe.scatter(_slytherin[y_name], _slytherin[x_name], s=2, color='green', alpha=0.5)
    axe.scatter(_ravenclaw[y_name], _ravenclaw[x_name], s=2, color='blue', alpha=0.5)
    axe.set_yticks([dtf[x_name].min(), (dtf[x_name].min() + dtf[x_name].max()) // 2, dtf[x_name].max()])
    axe.set_xticks([dtf[y_name].min(), (dtf[y_name].min() + dtf[y_name].max()) // 2, dtf[y_name].max()])
```

```python
dtf = csv_to_dataframe(path)
scatter_matrix_plot(dtf)
```

```python
scatter_matrix_plot(dtf, x_feature=['Astronomy', 'Defense Against the Dark Arts'], figsize=(20, 5))
```
