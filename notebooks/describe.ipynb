# Describe
This piece of code is used to have a quick look on the data.
Describe shows us, for each column, some basic info like the mean, the std, quartiles..
Before any analysis, knowing what data looks like is a great way to understand it and adapt formatting.

The code defines a script to describe a dataset, providing statistical summaries of its numerical columns. The main functionalities include:

1. **Command-Line Interface**:
   - The script uses `argparse` to handle command-line arguments, allowing users to specify the dataset file, and optional flags to drop empty columns (`-d` or `--drop`), print additional information (`-v` or `--verbose`), and include bonus information (`-b` or `--bonus`).

2. **Main Function (`describe`)**:
   - This function takes a DataFrame and optional flags as input and returns a DataFrame containing descriptive statistics of the numerical columns.
   - It can drop empty columns and include additional information like missing values and range if the `bonus` flag is set.

3. **Column Description (`describe_column`)**:
   - This helper function computes various statistics for a specific column, including count, mean, standard deviation, min, max, quantiles (25%, 50%, 75%), missing values, and range.

4. **Statistical Computations**:
   - Separate functions compute the mean (`compute_mean`), standard deviation (`compute_std`), min and max values (`find_min_max`), quantiles (`compute_quantile`), and total missing values (`get_total_missing_values`) for a column.

5. **Execution**:
   - The script reads a CSV file using a custom `open_csv` function from the `ft_dslr.tools` module.
   - It then describes the dataset based on the provided command-line arguments and prints the description.
   - If an error occurs during file opening or processing, it prints an error message.

The script is designed to be run from the command line, providing a flexible way to analyze and summarize datasets with optional verbosity and additional statistical information.
