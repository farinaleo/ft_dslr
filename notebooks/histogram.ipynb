# Histogram

This notebook is used to plot the histogram of the dataset. It is used to understand the distribution of the data.
The question is: Which Hogwarts course has a homogeneous score distribution between all four houses?

The code defines a script to plot histograms from a dataset, focusing on specific subjects and categorizing data by "Hogwarts House." The main functionalities include:

1. **Command-Line Interface**:
   - The script uses `argparse` to handle command-line arguments, allowing users to specify the dataset file, the subjects to plot (`--columns`), a flag to use all subjects (`--all`), and a verbosity flag (`-v` or `--verbose`).

2. **Main Function (`histogram`)**:
   - This function takes a DataFrame and optional parameters to plot histograms for specified subjects.
   - It drops the "Index" column and checks for the presence of "Hogwarts House" data.
   - It plots histograms for each subject, categorized by Hogwarts Houses, using different colors for each house.

3. **Plotting Function (`plot_single_histogram`)**:
   - This helper function plots a histogram for a single subject on a given axis, differentiating data by Hogwarts Houses with distinct colors.

4. **Execution**:
   - The script reads a CSV file using a custom `open_csv` function from the `ft_dslr.tools` module.
   - It then generates histograms based on the provided command-line arguments and displays the plot.
   - If an error occurs during file opening or processing, it prints an error message.

The script is designed to be run from the command line, providing a visual summary of dataset subjects categorized by Hogwarts Houses, with optional verbosity for additional information during execution.
